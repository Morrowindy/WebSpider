## 爬取概况

我们爬取的是2018年8月14日拉勾网“全国站”中“数据分析”相关的职位。

首先在拉勾网搜索“数据分析”，在“全国站”可以看到有30页结果。我们准备使用requests爬取搜索页面的内容，整理到excel表格中。  

由于拉勾网的搜索页面使用了Ajax技术，在不同页数之间切换时页面并没有刷新，使用chrome浏览器的开发工具可以观察到，页面切换时我们获得的其实是Type为xhr的Ajax请求。观察某个请求的详细情况，可以看到RequestURL、Cookie、User-Agent等信息，将其纳入爬虫的headers部分。同时将FormData的三个值作为data参数传入request.post()中。注意这里的cookies是直接复制粘贴的。  

Ajax请求返回的是json格式的数据，能够用python的字典方便地处理。观察Preview就能发现，这一页的15条数据就储存在‘content’-‘postionResult’-‘result’中，可以直接使用pandas转化为DataFrame。  

由于拉勾网的反爬虫较为严格而且我们手头没有代理，我们每爬取一页数据就将数据写入excel的一页sheet中，并暂停10-12秒。另外如果请求过快，服务器可能返回不含‘content’键值的json信息进行警告，因此我们使用try-except来捕捉因此引发的KeyError，并将已经爬取的数据保存下来。

```python
SLEEP_LOWER_TIME, SLEEP_UPPER_TIME = 10, 12
excelName = '拉勾网数据分析职位分页.xlsx'
```

```python
def get_one_page(pageNumber): # 爬取单页内容，返回json格式的数据
    url = 'https://www.lagou.com/jobs/positionAjax.json?needAddtionalResult=false'
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/66.0.3359.117 Safari/537.36',
        "Cookie": cookies, #填入自己浏览器中的cookies
        'Referer': 'https://www.lagou.com/jobs/list_%E6%95%B0%E6%8D%AE%E5%88%86%E6%9E%90?labelWords=&fromSearch=true&suginput='
    }
    post_data = {
        'first':'false',
        'pn':pageNumber,
        'kd': '数据分析'
    }
    
    response = requests.post(url, headers = headers, data = post_data)
    if response.status_code == 200:
        return response.json()
    return None
```

```python
def parse_one_page(jsonFile): #获取数据，转化为DataFrame，写入excel表格
    result = jsonFile['content']['positionResult']['result'] #获取职位相关的信息
    df = pd.DataFrame(result)
    time.sleep(random.uniform(SLEEP_LOWER_TIME, SLEEP_UPPER_TIME))
    return df
```

```python
def get_pages(start, end, file_name):# 爬取多页，并写入excel
    writer = pd.ExcelWriter(file_name)
    try:
        for i in range(start, end+1):
            jsonFile = get_one_page(i)
            df = parse_one_page(jsonFile)
            df.to_excel(writer, index=False, sheet_name=str(i))
            print('已保存第{}页'.format(i), end= ' ' if i%10 else '\n')
            time.sleep(12 + (i//10)*5)
    except KeyError:
        print('\n第{}页出现异常!'.format(i))
        writer.save()
    else:
        writer.save()
```

运行get_pages()函数，将这30页数据写入一个excel表格的30个sheet中。然后再用pandas将这30个sheet的数据合并。

```python
get_pages(1,30, excelName)
```

```python
def merge_sheets(sourceFile, newFile):
    writer = pd.ExcelWriter(newFile)
    df = pd.DataFrame()
    for i in range(30):
        sheet_df = pd.read_excel(sourceFile, sheet_name=str(i+1))
        df = df.append(sheet_df, ignore_index=True)
    df.to_excel(writer, index=False)
    writer.save()
    
merge_sheets('拉勾网数据分析职位分页.xlsx', '拉勾网数据分析职位.xlsx')
```

```python
def get_one_page(positionId):
    baseURL = 'https://www.lagou.com/jobs/{}.html'.format(positionId)
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/66.0.3359.117 Safari/537.36',
        "Cookie": cookies #同样在header加载cookies
    }
    response = requests.get(baseURL, headers=headers)
    if response.status_code == 200:
        return response.text
    print('Status_Code Error!')
    return None

def parse_one_page(html):
    html = etree.HTML(html)
    targetText = html.xpath('//dd[@class="job_bt"]//div/p/text()')
    description = ''.join(targetText).replace('\xa0', '')
    return description

def get_pages(idList):
    descriptionList = []
    for positionId in idList:
        try:
            html = get_one_page(positionId)
            description = parse_one_page(html)
            descriptionList.append(description)
            time.sleep(random.uniform(3,5))
        except:
            descriptionList.append('')
            time.sleep(60)
    return descriptionList

dfRaw['description'] = get_pages(dfRaw.positionId)
dfRaw.to_excel('拉勾网数据完整版.xlsx', index=False)
```

## 薪酬

网站给出的薪酬都是 “xk-xk”格式的区间，我们提取出每个职位的薪酬上下限，计算均值作为这个职位的薪酬均值。然后对薪酬均值、下限、上限的分布分别作图。  
可以看到，薪酬的分布并不符合正态分布，而是存在左偏。从均值看大部分职位集中在5K-10K左右，也有少数分布在20k以上，从下限看大部分职位的起薪是5k，但也存在起薪20k左右的高端职位，上限的参考价值较低，但也能看出大部分职位薪酬上限是10K，与均值反映的规律一致。  

总的说来，数据集中数据分析师岗位中大部分薪酬位于5-10K，也存在少量25k左右起步的职位，这可能反映的是初级分析师和高级分析师的差别。

```python
df['salaryLower'] = df.salary.str.extract(r'(\d+)k-.+', expand=True).astype('int').values
df['salaryUpper'] = df.salary.str.extract(r'.+-(\d+)k', expand=True).astype('int').values
df['salaryAvg'] = (df['salaryLower'] + df['salaryUpper'])/2

_ , ax = plt.subplots(ncols=3,figsize=(18,3))

sns.distplot(df.salaryAvg, ax=ax[0]) ; ax[0].set(title='薪酬均值分布', xlabel='千元/月')
sns.distplot(df.salaryLower, ax=ax[1]) ; ax[1].set(title='薪酬下限分布', xlabel='千元/月')
sns.distplot(df.salaryUpper, ax=ax[2]) ; ax[2].set(title='薪酬上限分布', xlabel='千元/月')
```

## 工具软件要求

那么用人单位对数据分析师的工具软件有什么要求呢？由于在职位要求中提到的这些软件基本上都使用英文名，我们首先提取出文本中的英文单词，移除无关的字词，再进行分析。  

```python
descString = df['description'].fillna(' ') #取出包含文本的一列

wordsSet = set()
pattern = '[a-zA-Z]+' 
for text in descString:
    result = re.findall(pattern, text) #查找单词
    result = (x.capitalize() for x in result) #统一为首字母大写
    wordsSet.update(result)

print(wordsSet)
```

从wordsSet中手动提取出工具软件相关的关键词。接下来将薪酬分为3梯队：5-10k，10-15k和15k以上，查看这些工具软件各自的分布，然后计算总数。

```python
##### 确定关键词
wordList = ['Scala', 'TensorFlow', 'Hadoop', 'Datamix',  'Python'
            , 'Word', 'Spark', 'SPSS', 'PPT',  'SQL', 'Hive', 'Excel'
            , 'R', 'SAS','Weka', 'Tableau', 'Powerpoint']
#首先将3个薪酬梯队的相应文本取出
group1 = descString.loc[df['salaryAvg']<=10]
group2 = descString.loc[df['salaryAvg'].map(lambda x: True if 10<x<=15 else False)]
group3 = descString.loc[df['salaryAvg']>15]

#进行数量统计
def count_by_wordList(wordList): #定义一个生成器，方便下一步转化为DataFrame
    for word in wordList:
        yield {
            '工具':word,
            '5k-10k':group1.str.contains(word, case=False).sum(), #统计在5-10k这个档次出现相应关键词描述的岗位个数，下同
            '10k-15k':group2.str.contains(word, case=False).sum(),
            '15k+':group3.str.contains(word, case=False).sum(),
        }
countDf = pd.DataFrame(count_by_wordList(wordList))[['工具', '5k-10k', '10k-15k', '15k+']]

#手动调整R和TF的值
pattern_R = r'[^a-zA-Z]R[^a-zA-Z]' #包含单独的大写数字R
pattern_TF = r'tf|tensor' #包含TF或tensor字样

for col, group in zip(['5k-10k', '10k-15k', '15k+'], [group1, group2, group3]):
    countDf.loc[countDf['工具']=='R', col] = group.str.contains(pattern_R, case=True).sum()
    countDf.loc[countDf['工具']=='TensorFlow', col] = group.str.contains(pattern_TF, case=False).sum()

countDf['sum'] = countDf['5k-10k'] + countDf['10k-15k'] +countDf['15k+'] #计算总量
countDf.sort_values(by='sum', ascending=False, inplace=True) #按总量排序
countDf.head(5) #查看总量最多的5个
```

可以看到，数据分析岗用人单位需求前五名的工具软件/语言是 Excel SQL Python R和SAS，下面作图看详细查看。

```python
plt.figure(figsize=(14,8))
sns.barplot(data=countDf, y='工具', x='sum', orient='h')
plt.xlabel('岗位数量', fontsize=12)
plt.ylabel('工具软件',fontsize=12)
plt.title('不同工具软件要求统计', fontsize=18)
plt.yticks(fontsize=14)
for index, num in enumerate(countDf['sum']):
    plt.text(s="%.1f%%"%(num/4.5), x=num+.4, y=index, fontsize=12)
```

从上图可以看到，用人单位对工具软件的要求按照比例可以分为3个层次：  
1. SQL和Excel 有60%以上的单位提到了这两个工具软件，说明对数据库或类数据库的操作是数据分析师的基本工作内容。
2. 提及Python,R,SPSS,SAS,Hive,Spark 这六项的的占比在25%-40%左右。前面4项是常用的数据分析语言/软件，后面两项是大数据相关的应用。这说明很多数据分析师岗位需要进行数据的分析挖掘处理，而且可能需要与大数据打交道。
3. 在20%及以下的软件种类较杂，包含了Office系列软件，其他数据可视化/数据挖掘工具（Tableau，Weka）和深度学习相关（TF）等。 

## 工具软件与薪酬

接下来我们看一看不同档次的薪酬对这些工具软件的需要是否存在差别。  

```python
_ , ax = plt.subplots(ncols=3,figsize=(17,6))

sns.barplot(data=countDf,x='5k-10k', y='工具', orient='h', ax=ax[0]) ; ax[0].set(title='薪酬5k-10k', xlabel='岗位数量')
sns.barplot(data=countDf,x='10k-15k', y='工具', orient='h',ax=ax[1]) ; ax[1].set(title='薪酬10k-15k', xlabel='岗位数量', ylabel='')
sns.barplot(data=countDf,x='15k+', y='工具', orient='h',ax=ax[2]) ; ax[2].set(title='薪酬15k+', xlabel='岗位数量', ylabel='')
```

上图反映的是不同薪酬档次中对工具软件需求的绝对数量，可以看到随着薪酬档次的提升，对大数据应用和Tableau等数据可视化工具的需要也在逐渐上升。下面的比例图可以更好地反映这一趋势。

```python
countDf_melt = pd.melt(countDf, id_vars='工具', value_vars=['5k-10k', '10k-15k', '15k+'], value_name='数量',var_name='薪酬组')
countDf_melt['薪酬组内岗位数'] = countDf_melt['薪酬组'].map({
    '5k-10k':len(group1), '10k-15k':len(group2), '15k+':len(group3)
})
countDf_melt['比例'] =  countDf_melt['数量']/countDf_melt['薪酬组内岗位数']

plt.figure(figsize=(18,6))
sns.barplot(data=countDf_melt, x='工具', y='比例', hue='薪酬组')
plt.legend(prop={'size':18})
plt.title('数据分析工具在不同薪酬组的比例', fontsize=20)
plt.xticks(fontsize=14)
plt.xlabel('')
plt.ylabel(s='比例', fontsize=15)
```

总的来说我们可以看出不同薪酬的岗位对数据分析工具有这样的差别和趋势：  
1. 无论哪个档次的薪酬，需要SQL和Excel的岗位都占据大多数。中端（10k-15k）薪酬岗位甚至100%要求SQL和Excel，高端薪酬岗位（15k+）对Excel的需求有所下降。
2. 四大数据分析语言/软件中，Python在高端薪酬岗位中的需求比例高达100%，而在中低端岗位相对较低。R、SPSS、SAS在中端薪酬岗位的比例最高，另外高端薪酬岗位没有提及SAS和SPSS。另外，中高端薪酬岗位对大数据应用Hive、Spark、Hadoop的要求比例也很高，Hive和Spark在高端薪酬岗位中甚至达到了100%。
3. Tableau、Weka等数据分析可视化软件只在高端薪酬岗位中有要求。

综上所述，我们可以大致推荐出一条数据分析师的进阶之路：  
入门级别的分析师主要从事数据的清洗整理，因此首先需要掌握Excel和SQL；中级的数据分析师还需要进一步挖掘数据进行建模，同时接触的数据量也大大增加，因此需要掌握Python等数据分析语言和Hive等大数据应用语言；进入高端后会涉及更多商务领域，此时就应当进一步使用Tableau等软件了。

# 第四部分：词云分析关键词

```python
jieba.add_word('大数据')
for w in "优先 掌握 数据 相关 熟悉 对数".split():
    jieba.del_word(w)
wordcloudSeries = (descString.str.replace('岗位职责','').str.replace('岗位要求', '') #进行分词
                   .apply(lambda x : jieba.analyse.extract_tags(x, topK=20, allowPOS=('n'))))#只提取名词

from collections import Counter
cnt = Counter()
for wl in wordcloudSeries: #使用Counter进行词频计数
    cnt.update(wl)

wc = (WordCloud(width=800, height=400, background_color='white'
                ,font_path='SimHei-windows.ttf')#必须在工作文件夹下有一个字体文件，否则中文不显示
      .generate_from_frequencies(cnt)) #传入Counter

plt.figure(figsize=(14,8))
plt.imshow(wc)
```
